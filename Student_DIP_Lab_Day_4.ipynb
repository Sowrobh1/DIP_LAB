a)	Write and execute programs to perform the following operations on images. At first, create one image with white background of size (500 X 500) and put a circle of black color on that image as foreground. Secondly, create another image with white background of size (500 X 500) and put a rectangle of red color as foreground.

•	Perform the logical operations (AND, OR and NOT) between two images


# Create first image (500x500) with white background and black circle

```python
image1 = np.ones((500, 500, 3), dtype=np.uint8) * 255
cv2.circle(image1, (250, 250), 100, (0, 0, 0), -1)
cv2_imshow(image1)
```

# Create second image (500x500) with white background and red rectangle

```python
image2 = np.ones((500, 500, 3), dtype=np.uint8) * 255
cv2.rectangle(image2,(100,150),(400,350),(0,0,255),-1) #BGR
cv2_imshow(image2)
```

# Addition

```python
addition = cv2.add(image1, image2)
cv2_imshow(addition)
```

# Subtraction

```python
subtraction = cv2.subtract(image1, image2)
cv2_imshow(subtraction)
```

# Perform multiplication operation

```python
multiplication = cv2.multiply(image1, image2)
cv2_imshow(multiplication)
```

# Convert images to grayscale for division (to prevent overflow issues)

With grayscale images, you only need to check for and handle zero values once, making the process more straightforward.

```python
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray1)
cv2_imshow(gray2)
```

# Division

```python
division = cv2.divide(gray1, gray2, dtype=cv2.CV_32F)
division = cv2.normalize(division, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2_imshow(division)
```

# Logical Operations

```python
bitwise_and = cv2.bitwise_and(image1, image2)
cv2_imshow(bitwise_and)
```

```python
bitwise_or = cv2.bitwise_or(image1, image2)
cv2_imshow(bitwise_or)
```

```python
bitwise_not1 = cv2.bitwise_not(image1)
cv2_imshow(bitwise_not1)
```

```python
bitwise_not2 = cv2.bitwise_not(image2) #negation
cv2_imshow(bitwise_not2)
```

```python
def show_images(images, titles):
    plt.figure(figsize=(10, 10))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(3, 3, i+1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(title)
        plt.axis("off")
    plt.show()

# Display results using matplotlib
show_images([
    image1, image2, addition, subtraction, multiplication,
    cv2.cvtColor(division, cv2.COLOR_GRAY2RGB), bitwise_and, bitwise_or, bitwise_not1
], [
    "Image 1 (Circle)", "Image 2 (Rectangle)", "Addition", "Subtraction", "Multiplication",
    "Division", "Bitwise AND", "Bitwise OR", "Bitwise NOT (Image 1)"
])
```
