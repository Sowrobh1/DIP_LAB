```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image_url = 'https://images.ctfassets.net/hrltx12pl8hq/3Z1N8LpxtXNQhBD5EnIg8X/975e2497dc598bb64fde390592ae1133/spring-images-min.jpg'
```

```python
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Real Image")
plt.show()
```

```python
resized_img = cv2.resize(img_rgb, (256, 256))
```

```python
rotated_img = cv2.rotate(img_rgb, cv2.ROTATE_180)
```

```python
flipped_img = cv2.flip(img_rgb, 0)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.imshow(resized_img)
plt.title("Resized")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(rotated_img)
plt.title("Rotated")
plt.axis('off')


plt.subplot(1, 4, 3)
plt.imshow(flipped_img)
plt.title("Flipped")
plt.axis('off')

plt.show()
```
